## DataSet Normalization
#### 1.Dimensional Reduction
#### PCA( principal component analysis)
dataSet includes features and target, based on information. Can not be trusted.

#### dimensional reduction
reduces the columns and make number on columns(features) lesser.
pros: 1. Lesser data 2.can be shown on diagram

#### Iris ---> 4features ----> 3features
the person who does this is called PCA.

```python
import pandas as pd
import matplotlib.pylab as plt

from sklearn import decomposition
from sklearn import preprocessing
iris= pd.read_csv("iris.csv")
# print(iris)
```

#### encodning:

```python
le=preprocessing.LabelEncoder()
iris.variety=le.fit_transform(iris.variety)#mapping
# print(iris)
```

#### Defining X, Y:

```python
X= iris.iloc[:,0:4].values
Y= iris.iloc[:,4].values
```

#### decomposition and fitting X:

```python
pca=decomposition.PCA(n_components=3)
pca.fit(X)
x=pca.transform(X)
# print(X)
# print("-"*50)
# print(x)
```

#### show data on diagram

```python
fig= plt.figure(figsize=(8,8))
ax= fig.add_subplot(111,projection='3d')

ax.scatter(x[Y==0,0],x[Y==0,1], x[Y==0,2],c="red")
ax.scatter(x[Y==1,0],x[Y==1,1], x[Y==1,2],c="green")
ax.scatter(x[Y==2,0],x[Y==2,1], x[Y==2,2],c="blue")
plt.show()
```
