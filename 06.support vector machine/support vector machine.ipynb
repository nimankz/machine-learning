#### support vector machine(SVM):
can find a line which, separate two groups.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
```

```python
x1=[1,8,1.5,8.7,1,9,2,10,2.3,9]
y1=[2,8,1.8,8.3,0.6,11,2,8.8,3.1,7.8]

plt.scatter(x1, y1,c="red")
plt.show()
```

```python
y=["A", "B","A", "B","A", "B","A", "B","A", "B"]
X=list(zip(x1, y1))
print(X)
print(y)
```

```python
model=svm.SVC(kernel="linear")
model.fit(X, y)
print(model.predict([[1,2]]))
print(model.predict([[11,12]]))
```

```python
xx=np.linspace(0,11,40)
print(len(xx))

w=model.coef_[0] #weight of groups
print(w[0])
print(w[1])
print(model.intercept_[0]) #angle(intercept)
```

```python
yy=((-w[0]/w[1]) * (xx) )- (model.intercept_[0]/w[1])
print(xx)
print(yy)
```

```python
# plt.scatter(xx,yy,c="blue")
plt.plot(xx,yy,c="blue")
plt.scatter(x1, y1,c="green")
plt.show()
```
